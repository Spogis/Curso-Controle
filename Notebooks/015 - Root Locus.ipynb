```python
import control
import matplotlib.pyplot as plt
```

```python
# Define transfer function of the system
s = control.tf('s')
```

## Root Locus

```python
sys =  ((s+5)*(s+3))/((s-2)*(s-4))
sys
```

```python
Poles = control.poles(sys)
Poles
```

```python
for n_poles in range(len(Poles)):
    print(Poles[n_poles])
```

```python
Zeros = control.zeros(sys)
Zeros
```

```python
roots, gains = control.root_locus_map(sys)
control.root_locus(sys)
```

```python
gm, pm, sm, wpc, wgc, wms = control.stability_margins(sys)
```

```python
#Gain margin
gm
```

```python
#Phase margin
pm
```

```python
#Stability margin, the minimum distance from the Nyquist plot to -1
sm
```

```python
#Phase crossover frequency (where phase crosses -180 degrees), which is associated with the gain margin.
wpc
```

```python
#Gain crossover frequency (where gain crosses 1), which is associated with the phase margin.
wgc
```

```python
#Stability margin frequency (where Nyquist plot is closest to -1)
wms
```
